```python
import os
import sqlalchemy
SQLALCHEMY_SILENCE_UBER_WARNING = 1
import pandas as pd
```

```python
beers_path = "https://raw.githubusercontent.com/sugus-labs/data-science-the-bridge/main/2-data-analysis/2-Pandas/Ejercicios/datasets_beers.csv"
breweries_path = "https://raw.githubusercontent.com/sugus-labs/data-science-the-bridge/main/2-data-analysis/2-Pandas/Ejercicios/datasets_breweries.csv"
```

```python
def connect_tcp_socket() -> sqlalchemy.engine.base.Engine:
    """ Initializes a TCP connection pool for a Cloud SQL instance of MySQL. """
    db_host = '34.132.212.141'
    db_user = 'root'
    db_pass = '123456'
    db_name = 'main'
    db_port = '3306'

    engine = sqlalchemy.create_engine(
        sqlalchemy.engine.url.URL.create(
            drivername="mysql+pymysql",
            username=db_user,
            password=db_pass,
            host=db_host,
            port=db_port,
            database=db_name,
        ),
    )
    return engine
```

```python
engine = connect_tcp_socket()
conn = engine.connect()
```

```python
result = conn.execute("SHOW TABLES;").fetchall()
for r in result:
    print(r)
```
